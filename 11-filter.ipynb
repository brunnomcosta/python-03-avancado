Filter:
- É uusada para filtrar elementos de um iterável com base em uma função que pode ter True ou False.
- Objetifo é permitir apenas os elementos que satisfazem uma condição

```python
def e_par(x):
    return x % 2 == 0

numeros = [1, 2, 3, 4, 5]

numeros_pares = filter(e_par, numeros)
list(numeros_pares)

# Resultado esperado:
# [2, 4]
```

```python
produtos = [
    {'nome': 'Produto A', 'preco': 10.0, 'em_estoque': True},
    {'nome': 'Produto B', 'preco': 20.0, 'em_estoque': False},
    {'nome': 'Produto C', 'preco': 30.0, 'em_estoque': True}
]

def produto_em_estoque(produto):
    return produto['em_estoque']

produtos_em_estoque = filter(produto_em_estoque, produtos)
list(produtos_em_estoque)

# Resultado esperado:
# [{'nome': 'Produto A', 'preco': 10.0, 'em_estoque': True},
#  {'nome': 'Produto C', 'preco': 30.0, 'em_estoque': True}]
```

```python
transacoes = [200.0, -50.0, 300.0, -100.0, 150.0]

def transacao_despesa(valor):
    return valor < 0

despesas = filter(transacao_despesa, transacoes)
list(despesas)

# Resultado esperado:
# [-50.0, -100.0]
```
