# Geradores e `yield`

`yield` e a palavra-chave usada para criar geradores em Python. Quando ela e
utilizada dentro de uma funcao, essa funcao passa a produzir valores aos poucos,
em vez de retornar tudo de uma vez. Um gerador e um tipo especial de iterador.

A diferenca principal entre `return` e `yield` esta no comportamento da execucao:

- `return`: encerra a funcao imediatamente e devolve um unico resultado.
- `yield`: pausa a funcao, devolve um valor e permite continuar depois do mesmo ponto.

Geradores sao importantes porque permitem processar sequencias de forma mais
eficiente em memoria, ja que os valores podem ser produzidos sob demanda.
Isso e especialmente util quando a sequencia e grande ou potencialmente infinita.

No exemplo abaixo, a funcao gera apenas os numeros impares ate o limite informado.

```python
# Gerador que produz numeros impares de 1 ate `n`.
def numeros_impares(n):
    """Gera numeros impares ate o limite informado."""

    for i in range(1, n + 1, 2):
        yield i
```

```python
# O `for` consome o gerador e solicita um valor por vez.
for numero in numeros_impares(10):
    print(numero)


# Resultado esperado:
# 1 -> primeiro numero impar ate 10.
# 3 -> proximo impar gerado.
# 5 -> segue pulando os pares.
# 7 -> novo valor impar da sequencia.
# 9 -> ultimo impar antes de 10.
```
