# `enumerate` — Iteração com Índice e Valor

A função built-in `enumerate()` retorna um objeto iterável que produz tuplas `(índice, valor)` para cada elemento de uma sequência.

**Sintaxe:**
```python
enumerate(iterável, start=0)
```

- `iterável`: qualquer sequência (lista, tupla, string, etc.)
- `start` *(opcional)*: valor inicial do contador de índice (padrão: `0`)

**Quando usar:**
- Quando você precisa do índice **e** do valor ao iterar uma lista
- Substitui o padrão manual `for i in range(len(lista))`

````python
frutas = ["maçã", "banana", "laranja"]

for fruta in enumerate(frutas):
    print(f"Índice: {fruta[0]}, Fruta: {fruta[1]}")
````

## Exemplo 1 — Iterando com índice usando `for`

O `enumerate` retorna uma tupla `(índice, valor)` a cada iteração. Podemos acessar os elementos pela posição (`fruta[0]`, `fruta[1]`) ou usar **desempacotamento** direto: `for indice, fruta in enumerate(frutas)`.

````python
numeros = [10, 20, 30, 45, 50, 55, 60]

indices_valores = {indice: valor for indice, valor in enumerate(numeros)}
indices_valores
````

## Exemplo 2 — Criando dicionário `{índice: valor}` com compreensão

`enumerate` pode ser combinado com **dict comprehension** para gerar um dicionário que mapeia cada índice ao seu valor correspondente na lista.

````python
indices_valores.items()
````

## Exemplo 3 — Visualizando os pares com `.items()`

O método `.items()` de um dicionário retorna uma view com todos os pares `(chave, valor)`, útil para inspecionar ou iterar o dicionário.

````python
indices_pares = {indice: valor for indice, valor in indices_valores.items() if indice % 2 == 0}
indices_pares
````

## Exemplo 4 — Filtrando por índice par com dict comprehension

Combinamos `.items()` com uma condição `if` para selecionar apenas as entradas cujo **índice é par** (`indice % 2 == 0`), demonstrando como filtrar um dicionário pelo índice de origem.
