# Walrus Operator (`:=`)

O operador walrus permite atribuir um valor a uma variavel e usar esse valor
na mesma expressao. Ele e util quando voce quer deixar o codigo mais direto
sem repetir a mesma operacao.

## Exemplo 1

Neste caso, o valor atual da lista e salvo em `maior` dentro da condicao.
Depois disso, o mesmo valor ja pode ser usado para decidir se ele entra no resultado.

```python
numeros = [5, 12, 3, 18, 7, 20]

# Guarda o valor atual em `maior` e filtra apenas os numeros acima de 10.
numeros_maiores_que_10 = [n for n in numeros if (maior := n) > 10]
print(numeros_maiores_que_10)

# Resultado esperado:
# [12, 18, 20] -> apenas os valores maiores que 10 permanecem na lista final.
```

## Exemplo 2

Aqui o walrus e usado para limpar cada linha do arquivo com `strip()` e,
ao mesmo tempo, testar se a linha ainda possui conteudo antes de adicionar na lista.

```python
linhas_processadas = []

with open("05-texto.txt", "r", encoding="utf-8") as arquivo:
    for linha in arquivo:
        # `linha_limpa` recebe o texto sem espacos extras.
        if (linha_limpa := linha.strip()):
            linhas_processadas.append(linha_limpa)

print(linhas_processadas)

# Resultado esperado:
# ['Ol? mundo', 'Esta ? uma linha de teste', 'Python ? f?cil', 'Ruby ? interessante']
# -> a lista final contem apenas linhas com texto; linhas vazias sao descartadas pelo `if`.
```
