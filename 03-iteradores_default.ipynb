# Iteradores padrao do Python

Antes de criar iteradores personalizados, vale entender que varios objetos do
Python ja sabem ser percorridos naturalmente. Listas, dicionarios, `range()`
e utilitarios como `zip()` e `itertools` seguem o protocolo de iteracao.

Isso e importante porque o comando `for` usa esse protocolo internamente por
meio de `iter()` e `next()`. Na pratica, iteradores permitem percorrer dados
de forma padronizada, reutilizavel e, em muitos casos, mais eficiente em
memoria, pois os valores podem ser produzidos sob demanda.

```python
# Exemplo 1 - `range()` gera numeros sob demanda.
# Aqui comecamos em 1, vamos ate 5 e pulamos de 2 em 2.
for numero in range(1, 6, 2):
    print(numero)


# Resultado esperado:
# 1 -> primeiro numero da sequencia.
# 3 -> soma 2 ao valor anterior.
# 5 -> ultimo valor antes de ultrapassar o limite final `6`.
```

```python
# Exemplo 2 - Listas ja sao iteraveis por padrao.
frutas = ['maca', 'banana', 'laranja']
for fruta in frutas:
    print(f'Fruta encontrada: {fruta}')


# Resultado esperado:
# Fruta encontrada: maca     -> primeiro item da lista.
# Fruta encontrada: banana   -> segundo item da lista.
# Fruta encontrada: laranja  -> terceiro e ultimo item.
```

```python
# Exemplo 3 - Dicionarios podem ser percorridos por chave e valor.
pessoa = {'nome': 'Brunno', 'idade': 30, 'cidade': 'Sao Paulo'}
for chave, valor in pessoa.items():
    print(f'{chave.title()}: {valor}')


# Resultado esperado:
# Nome: Brunno      -> mostra o valor associado a chave `nome`.
# Idade: 30         -> mostra o valor associado a chave `idade`.
# Cidade: Sao Paulo -> mostra o valor associado a chave `cidade`.
```

```python
# Exemplo 4 - `enumerate()` combina indice e valor.
tarefas = ['estudar', 'praticar', 'revisar']
for posicao, tarefa in enumerate(tarefas, start=1):
    print(f'{posicao}. {tarefa}')


# Resultado esperado:
# 1. estudar  -> primeira tarefa com indice iniciado em 1.
# 2. praticar -> segunda tarefa da lista.
# 3. revisar  -> terceira tarefa da lista.
```

```python
# Exemplo 5 - `zip()` percorre varias sequencias em paralelo.
nomes = ['Brunno', 'Maria', 'Joao']
idades = [30, 25, 28]
cidades = ['Sao Paulo', 'Rio de Janeiro', 'Curitiba']

for nome, idade, cidade in zip(nomes, idades, cidades):
    print(f'{nome} tem {idade} anos e mora em {cidade}')


# Resultado esperado:
# Brunno tem 30 anos e mora em Sao Paulo         -> primeira combinacao das tres listas.
# Maria tem 25 anos e mora em Rio de Janeiro     -> segunda combinacao.
# Joao tem 28 anos e mora em Curitiba            -> terceira e ultima combinacao.
```

```python
# Exemplo 6 - `itertools.count()` cria uma contagem potencialmente infinita.
# O `break` limita a demonstracao para nao gerar um loop sem fim.
import itertools

for numero in itertools.count(start=10, step=2):
    if numero > 20:
        break
    print(numero)


# Resultado esperado:
# 10 -> valor inicial definido em `start=10`.
# 12 -> soma 2 ao valor anterior.
# 14 -> continua seguindo o `step=2`.
# 16 -> proximo numero da contagem.
# 18 -> penultimo valor exibido.
# 20 -> ultimo valor aceito antes do `break`.
```
