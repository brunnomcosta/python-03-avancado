# Programação Funcional

Paradigma que trata a computação como a **avaliação de funções matemáticas**, evitando estados mutáveis e efeitos colaterais.

**Princípios fundamentais:**
- **Imutabilidade** — dados não são modificados, novos valores são criados
- **Funções puras** — o resultado depende apenas dos argumentos; sem efeitos colaterais
- **Composição** — funções complexas são construídas a partir de funções simples

**Ferramentas nativas do Python:**

| Recurso | Descrição |
|---|---|
| Funções de alta ordem | Funções que recebem ou retornam outras funções |
| `map` | Aplica uma função a cada elemento de um iterável |
| `filter` | Filtra elementos de um iterável com base em uma condição |
| `lambda` | Define funções anônimas de forma concisa |
| `lru_cache` | Memoiza resultados de funções para evitar recomputação |

## Funções de Alta Ordem

Uma **função de alta ordem** é aquela que recebe outra função como argumento ou a retorna como resultado.

No exemplo abaixo, `aplicar_funcao` recebe `quadrado` como argumento e a aplica sobre cada elemento da lista — sem precisar saber o que a função faz internamente.

```python
def quadrado(x):
    """Retorna o quadrado de x."""
    return x * x


def aplicar_funcao(funcao, lista):
    """
    Aplica uma função a cada elemento de uma lista (função de alta ordem).

    Args:
        funcao: Qualquer callable que aceite um único argumento.
        lista:  Iterável de valores a serem transformados.

    Returns:
        Lista com os resultados de funcao(x) para cada x em lista.
    """
    return [funcao(x) for x in lista]
```

```python
numeros = [1, 2, 3, 4, 5]
resultado = aplicar_funcao(quadrado, numeros)
print(resultado)
```

```python
usuarios = [
    {"nome": "Alice", "idade": 30},
    {"nome": "Bob", "idade": 25},
    {"nome": "Charlie", "idade": 35},
    {"nome": "David", "idade": 28}
]

def extrair_nome(usuarios):
    return list(map(lambda u: u["nome"], usuarios))

def filtrar_maiores_que(usuarios, idade):
    return list(filter(lambda u: u["idade"] > idade, usuarios))

def calcular_media_idade(usuarios):
    
    if not usuarios: 
        return 0
    
    idades = list(map(lambda u: u["idade"], usuarios))

    return sum(idades) / len(usuarios)

usuarios_maiores_que_27 = filtrar_maiores_que(usuarios, 27)
print("Usuários maiores que 27 anos:", extrair_nome(usuarios_maiores_que_27))

nomes_maior_que_27 = extrair_nome(usuarios_maiores_que_27)
print("Nomes dos usuários maiores que 27 anos:", nomes_maior_que_27)

media_idade = calcular_media_idade(usuarios_maiores_que_27)
print("Média de idade dos usuários maiores que 27 anos:", media_idade)
```

## map, filter e lambda

`map` e `filter` são funções de alta ordem embutidas no Python.  
`lambda` cria funções anônimas de uma linha, muito usadas como argumento para essas funções.

- `map(fn, iterable)` — transforma cada elemento
- `filter(fn, iterable)` — mantém apenas os elementos onde `fn` retorna `True`

O exemplo abaixo combina as duas para extrair nomes e calcular a média de idade de usuários filtrados por critério.

```python
from functools import lru_cache


@lru_cache(maxsize=None)
def fibonacci(n):
    """
    Retorna o n-ésimo número da sequência de Fibonacci usando memoização.

    Args:
        n: Índice da sequência (0-indexado). Deve ser >= 0.

    Returns:
        O n-ésimo número de Fibonacci.
    """
    if n <= 1:
        return n
    return fibonacci(n - 1) + fibonacci(n - 2)


for i in range(10):
    print(f"fibonacci({i}) = {fibonacci(i)}")
```

## Memoização com `lru_cache`

**Memoização** é uma técnica de otimização que armazena o resultado de chamadas anteriores de uma função para evitar recomputação.

`functools.lru_cache` implementa isso como decorator:
- `maxsize=None` — cache ilimitado (armazena todos os resultados já calculados)
- Ideal para funções **recursivas puras**, como Fibonacci, onde os mesmos valores são recalculados repetidamente

Sem cache, `fibonacci(n)` teria complexidade **O(2ⁿ)**. Com cache, cai para **O(n)**.
