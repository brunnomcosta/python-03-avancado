# `zip()` — Combinando Iteráveis

A função built-in `zip()` combina múltiplos iteráveis (listas, tuplas, etc.) em um único iterador de tuplas.

**Sintaxe:**
```python
zip(iterável1, iterável2, ...)
```

**Comportamento importante:**
- Retorna um **iterador** (não uma lista), que pode ser consumido uma única vez.
- Para quando o **iterável mais curto** se esgota (comportamento padrão).
- Para preservar todos os elementos do maior iterável, use `itertools.zip_longest()`.


## Exemplo 1 — Iterando sobre dois iteráveis em paralelo

`zip()` pareia os elementos de `nomes` e `idades` pelo índice, permitindo iterar sobre ambos simultaneamente.

````python
nomes = ["Alice", "Bob", "Charlie", "Diana"]
idades = [25, 30, 35, 40]

for nome, idade in zip(nomes, idades):
    print(f"{nome} tem {idade} anos.") 
````

## Exemplo 2 — Convertendo para dicionário com `dict(zip())`

Combinando `dict()` e `zip()` é possível criar um dicionário a partir de duas listas: uma de chaves e outra de valores.

````python
pessoas = dict(zip(nomes, idades ))
print(pessoas)
````

## Exemplo 3 — Filtrando o dicionário com dict comprehension

Com o dicionário criado, é possível aplicar um filtro usando **dict comprehension** para selecionar apenas os pares que atendem a uma condição.

````python
pessoas_acima_de_30 = {nome: idade for nome, idade in pessoas.items() if idade > 30}
print(pessoas_acima_de_30)  
````
