# Iterador com filtro por letra inicial

Neste exemplo, a classe `NomeComLetra` percorre uma lista de nomes e retorna
apenas aqueles que comecam com a letra informada. O estado interno e mantido
por meio do indice atual da lista.

Os iteradores sao importantes no Python porque definem a forma padrao de
percorrer colecoes e objetos personalizados. Isso torna o codigo compativel
com loops, filtros e outras operacoes da linguagem, alem de favorecer
solucoes mais eficientes e reutilizaveis.

```python
# Percorre a lista e devolve somente os nomes que comecam com a letra desejada.
class NomeComLetra:
    """Iterador que filtra nomes pela letra inicial."""

    def __init__(self, nomes, letra):
        # Colecao original que sera percorrida.
        self.nomes = nomes
        # Letra usada como criterio de filtro.
        self.letra = letra
        # Controla a posicao atual durante a iteracao.
        self.index = 0

    def __iter__(self):
        # O objeto atende ao protocolo de iteracao retornando a si mesmo.
        return self
    
    def __next__(self):
        # Avanca pela lista ate encontrar um nome compativel.
        while self.index < len(self.nomes):
            nome = self.nomes[self.index]
            self.index += 1
            if nome.startswith(self.letra):
                return nome
        # Se nenhum outro item atender ao filtro, a iteracao termina.
        raise StopIteration
    
# Dados de exemplo para demonstrar o funcionamento do iterador.
lista_nomes = ["Alice", "Bob", "Charlie", "David", "Eve"]
letra = "A"

# O loop `for` chama `iter()` e `next()` automaticamente.
nomes_com_a = NomeComLetra(lista_nomes, letra)
for nome in nomes_com_a:
    print(nome)


# Resultado esperado:
# Alice -> apenas este nome comeca com a letra `A`; os demais sao ignorados pelo filtro.
```
