# Exemplo de gerador lendo arquivo

Este notebook mostra um uso pratico de geradores: percorrer um arquivo texto
linha por linha e retornar apenas as linhas que atendem a uma regra.

A funcao `linhas_com_numeros_de_palavras()` nao carrega todas as linhas na
memoria ao mesmo tempo. Em vez disso, ela produz uma linha por vez com `yield`,
o que torna a abordagem mais eficiente para arquivos grandes.

Neste exemplo, o filtro usado e simples: a linha so e devolvida quando possui
exatamente a quantidade de palavras informada.

```python
# Gerador que filtra linhas de um arquivo pela quantidade de palavras.
def linhas_com_numeros_de_palavras(nome_arquivo, numero_palavras):
    """Gera linhas que tenham exatamente `numero_palavras` palavras."""

    with open(nome_arquivo, 'r', encoding='utf-8') as file:
        for linha in file:
            palavras = linha.split()
            if len(palavras) == numero_palavras:
                yield linha.strip()
```

```python
# O loop consome o gerador e imprime apenas as linhas com 3 palavras.
arquivo = '05-texto.txt'
numero_palavras = 3

for linha in linhas_com_numeros_de_palavras(arquivo, numero_palavras):
    print(linha)


# Resultado esperado:
# Python ? f?cil       -> linha do arquivo que possui exatamente 3 palavras.
# Ruby ? interessante  -> outra linha com exatamente 3 palavras.
```
