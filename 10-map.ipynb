# Map

A funcao `map` aplica uma funcao a cada item de um iteravel.
Ela e util quando voce quer transformar varios dados seguindo a mesma regra.

## Como usar

A estrutura basica e:

`map(funcao, iteravel)`

O resultado de `map` e um iterador. Por isso, normalmente convertemos para `list`
quando queremos visualizar todos os valores de uma vez.

```python
def quadrado(x):
    return x ** 2

numeros = [1, 2, 3, 4, 5]
resultados = map(quadrado, numeros)
print(list(resultados))


# Resultado esperado:
# [1, 4, 9, 16, 25] -> lista com o quadrado de cada numero da entrada.
```

## Exemplo com `lambda`

Quando a transformacao e pequena, `lambda` pode deixar o codigo mais direto.

```python
precos = [100, 250, 80, 45]
precos_com_desconto = map(lambda preco: round(preco * 0.9, 2), precos)
print(list(precos_com_desconto))


# Resultado esperado:
# [90.0, 225.0, 72.0, 40.5] -> cada preco recebeu 10% de desconto.
```

## Convertendo temperaturas

Outro uso comum e converter valores de uma escala para outra.

```python
def celsius_para_fahrenheit(c):
    return (c * 9 / 5) + 32


temperaturas_celsius = [0, 10, 20, 30, 40]
temperaturas_fahrenheit = map(celsius_para_fahrenheit, temperaturas_celsius)
print(list(temperaturas_fahrenheit))


# Resultado esperado:
# [32.0, 50.0, 68.0, 86.0, 104.0] -> conversao de cada valor de Celsius para Fahrenheit.
```

## Limpando textos

`map` tambem e util para padronizar strings, como remover espacos extras
e deixar tudo em minusculo.

```python
def normalizar_nome(nome):
    return nome.strip().lower()


nomes = ["  Alice  ", "Bob", "  Charlie  "]
nomes_normalizados = map(normalizar_nome, nomes)
print(list(nomes_normalizados))


# Resultado esperado:
# ['alice', 'bob', 'charlie'] -> nomes sem espacos extras e todos em minusculas.
```

## `map` com dois iteraveis

Tambem e possivel usar mais de um iteravel ao mesmo tempo.
Nesse caso, a funcao recebe um argumento para cada lista.

```python
nomes = ["Ana", "Bruno", "Carla"]
sobrenomes = ["Silva", "Costa", "Souza"]

nomes_completos = map(lambda nome, sobrenome: f"{nome} {sobrenome}", nomes, sobrenomes)
print(list(nomes_completos))


# Resultado esperado:
# ['Ana Silva', 'Bruno Costa', 'Carla Souza'] -> combinacao de cada nome com o sobrenome da mesma posicao.
```

## Resumo

- Use `map` quando a mesma transformacao precisa ser aplicada a varios itens.
- Se precisar visualizar o resultado, converta para `list`.
- Para operacoes simples, `lambda` pode ajudar a deixar o exemplo menor.
