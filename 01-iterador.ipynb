# Exemplo de iterador personalizado

Este notebook mostra como implementar o protocolo de iteracao em Python.
A classe `Contador` produz valores de `1` ate o limite informado, encerrando
a iteracao com `StopIteration`.

Os iteradores sao importantes no Python porque sustentam estruturas como
`for`, `next()`, compreensoes e varias funcoes nativas. Eles permitem
processar dados sob demanda, mantendo o estado da iteracao e evitando criar
toda a sequencia na memoria de uma vez.

```python
# Implementa um iterador manual que conta de 1 ate `limite`.
class Contador:
    """Retorna numeros inteiros em sequencia, comecando em 1."""

    def __init__(self, limite):
        # Define o valor maximo que sera produzido pelo iterador.
        self.limite = limite
        # Mantem o estado interno entre uma chamada e outra de `next()`.
        self.atual = 1

    def __iter__(self):
        # O proprio objeto atua como iterador.
        return self

    def __next__(self):
        # Encerra a iteracao quando o limite for ultrapassado.
        if self.atual > self.limite:
            raise StopIteration
        else:
            # Guarda o valor atual, avanca o estado e devolve o numero.
            valor = self.atual
            self.atual += 1
            return valor


# Consome o iterador em um loop `for`.
contador = Contador(10)
for numero in contador:
    print(numero)


# Resultado esperado:
# 1   -> primeira chamada de `next()` devolve 1.
# 2   -> segunda chamada devolve 2.
# 3   -> o iterador continua avancando de 1 em 1.
# 4   -> novo valor gerado pelo estado interno `self.atual`.
# 5   -> metade do intervalo solicitado.
# 6   -> a contagem segue em ordem crescente.
# 7   -> o loop continua enquanto nao houver `StopIteration`.
# 8   -> mais um valor produzido pelo iterador.
# 9   -> penultimo numero antes do limite.
# 10  -> ultimo valor permitido; depois disso a iteracao termina.
```
