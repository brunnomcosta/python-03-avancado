# Type Hints em Python

**Type Hints** (ou anotações de tipo) são uma forma de indicar explicitamente o tipo esperado de variáveis, parâmetros e valores de retorno de funções.

## Por que usar Type Hints?

- **Legibilidade**: o código fica mais fácil de entender
- **Detecção de erros**: ferramentas como `mypy` identificam erros de tipo antes da execução
- **Autocompletar**: IDEs usam as anotações para oferecer sugestões mais precisas
- **Documentação viva**: as anotações servem como documentação inline do código

> **Importante:** Type Hints são apenas indicativos — o Python **não** impede a execução se um tipo errado for passado em tempo de execução.

## Sintaxe básica

A sintaxe usa `:` para parâmetros e variáveis, e `->` para o tipo de retorno de funções:

```python
def funcao(parametro: tipo) -> tipo_retorno:
    ...
```

### Tipos primitivos mais comuns

| Tipo Python | Anotação |
|-------------|----------|
| Inteiro     | `int`    |
| Decimal     | `float`  |
| Texto       | `str`    |
| Booleano    | `bool`   |
| Sem retorno | `None`   |

No exemplo abaixo, `soma` recebe dois `int` e retorna um `int`:

````python
def soma(a: int, b: int) -> int:
    return a + b

print(soma(2, 3))
print(soma(5, 7))
print(soma(10, 15))
````

## Tipos compostos com `typing`

Para tipos mais complexos como listas, dicionários e tuplas, usamos o módulo `typing` (Python 3.8 ou anterior) ou diretamente os tipos nativos (Python 3.9+):

| Tipo              | `typing` (3.8-)      | Nativo (3.9+)      |
|-------------------|----------------------|--------------------|
| Lista de inteiros | `List[int]`          | `list[int]`        |
| Dicionário        | `Dict[str, int]`     | `dict[str, int]`   |
| Tupla             | `Tuple[int, str]`    | `tuple[int, str]`  |
| Opcional          | `Optional[str]`      | `str \| None`      |

No exemplo abaixo, criamos uma classe `Estudante` com anotações em todos os atributos e métodos, e uma função que filtra estudantes com média acima de um valor mínimo:

````python
from typing import List

class Estudante:
    def __init__(self, nome: str, idade: int, notas: List[float]) -> None:
        self.nome = nome
        self.idade = idade
        self.notas = notas

    def calcular_media(self) -> float:
        return sum(self.notas) / len(self.notas)
    
def alunos_acima_de_media(estudantes: List[Estudante], media_minima: float) -> List[str]:
    return [estudante.nome for estudante in estudantes if estudante.calcular_media() > media_minima]

estudandes = [
    Estudante("Alice", 20, [8.5, 9.0, 7.5]),
    Estudante("Bob", 22, [6.0, 7.0, 5.5]),
    Estudante("Charlie", 19, [9.0, 9.5, 8.0])
]

resultados = alunos_acima_de_media(estudandes, 8.0)
print(resultados)
````

## Resumo

- Use `: tipo` para anotar parâmetros e variáveis
- Use `-> tipo` para indicar o retorno de funções
- Use `List[tipo]`, `Dict[chave, valor]`, etc. do módulo `typing` para coleções (Python 3.8-)
- Em Python 3.9+, você pode usar `list[tipo]`, `dict[chave, valor]` diretamente sem importar `typing`
- Type Hints **não alteram o comportamento** do programa — são apenas informações para ferramentas e desenvolvedores
